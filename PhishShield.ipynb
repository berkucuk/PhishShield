```python
import os
import shutil
import numpy as np

def move_html_files(source_directory, destination_directory):
    # Kaynak dizindeki tm dosyalar kontrol et
    for filename in os.listdir(source_directory):
        if filename.endswith(".htm") or filename.endswith(".html"):
            source_path = os.path.join(source_directory, filename)
            destination_path = os.path.join(destination_directory, filename)
            # Dosyay hedef dizine ta
            shutil.move(source_path, destination_path)
            print(f"{filename} dosyas tand.")

# Kaynak dizin ve hedef dizini burada belirleyin
dir = "/content/Phishing_dataset/"
date = ["2018/","2019/","2020/"]
typ = ["malicious/","benign/"]
# Dosyalarn tanaca fonksiyonu ar
for t in typ:
  ndir = dir+t
  for d in date:
    source_dir = ndir + d + d
    destination_dir = ndir
    print(source_dir)
    move_html_files(source_dir, destination_dir)
```

```python
import os
import pandas as pd
import chardet
import codecs

def read_html_files(directory, label):
    data = []
    for filename in os.listdir(directory):
        if filename.endswith(".htm") or filename.endswith(".html"):
            file_path = os.path.join(directory, filename)
            try:
                with codecs.open(file_path, "r", encoding="utf-8", errors="ignore") as file:
                    content = file.read()
                    if content:  # Bo ierii olan dosyalar atla
                        data.append((content, label))
            except UnicodeDecodeError:
                print(f"UnicodeDecodeError: Dosya: {file_path}")
    return data

# Bening dosyalar oku ve class kolonunu 0 olarak etiketle
bening_directory = "/content/Phishing_dataset/benign"
bening_data = read_html_files(bening_directory, 0)

# Malicious dosyalar oku ve class kolonunu 1 olarak etiketle
malicious_directory = "/content/Phishing_dataset/malicious"
malicious_data = read_html_files(malicious_directory, 1)

# DataFrame'i olutur
data = bening_data + malicious_data
df = pd.DataFrame(data, columns=["content", "class"])

# DataFrame'i kontrol et
print(df.head())
```

```python
idx = tokenizer.word_index
inverse_map = dict(zip(idx.values(), idx.keys()))
```

```python
def tokens_to_string(tokens):
    words = [inverse_map[token] for token in tokens if token!=0]
    text = ' '.join(words)
    return text
```

```python
model = Sequential()
```

```python
embedding_size = 100
```

```python
model.add(Embedding(input_dim=num_words,
                    output_dim=embedding_size,
                    input_length=max_tokens,
                    name='embedding_layer'))
```

```python
model.add(GRU(units=16, return_sequences=True))
model.add(GRU(units=8, return_sequences=True))
model.add(GRU(units=4))
model.add(Dense(1, activation='sigmoid'))
```

```python
optimizer = Adam(learning_rate=1e-3)
```

```python
from keras.losses import binary_crossentropy
lss = binary_crossentropy
```

```python
model.compile(optimizer= 'adam' , loss= lss, metrics=['accuracy'])
```
